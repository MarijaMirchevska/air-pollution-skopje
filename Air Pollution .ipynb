## Data cleaning and restructuring
Because the mobile column is from a mobile monitoring station located in the municipality of Butel and occasionally displaced, we can drop that column.

We can also rename the column time to datetime and create additional separate time and date columns. 

We can add a column in each of the dataframes indicating the type of pollutant and concatenate the dataframes.

```python
# Data cleaning and restructuring

# Adding the pollutant column
dfs = [co, no2, o3, pm10, pm25] 
pollutants = ['CO', 'NO2', 'O3', 'PM10', 'PM25'] 
#in the code below when using pollutants it will refere to this list (q1,q2,q3,q4,q5)

for df, pollutant in zip(dfs, pollutants):
    df['Pollutant'] = pollutant
co
```

```python
# Renaming column time in datetime and dropping duplicates
for df in dfs:
    df.rename(columns = {'time' : 'Datetime'}, inplace=True)
    df.drop_duplicates(subset=['Datetime'], keep='first', inplace=True)
co
```

```python
# Concatenating the dataframes and 
df = pd.concat([co, no2, o3, pm10, pm25], ignore_index=True)

# Adding two new columns date and time

df['Date'] = df['Datetime'].dt.date
df['Time'] = df['Datetime'].dt.time

# Deleting unnecessary columns

df.drop(['Unnamed: 0.1', 'Unnamed: 0','Mobile','Datetime'], axis = 1, inplace=True)
print(df)
```

What location in Skopje has the highest average pollution for each of the pollutants? .......... (2 points)

# Question 2
For every year, for each of the pollutants, on what time stamp is the maximum pollution for every location? What time of the year seems to have the most pollution in Skopje? .......... (2 points)

```python
#Solution

df['Year'] = pd.to_datetime(df['Date']).dt.year
#in the code below this column is used in q4,q6,q7

result=[]
for year in df['Year'].unique():
    year_df = df[df['Year'] == year]
    
    for pollutant in year_df['Pollutant'].unique():
        pollutant_df = year_df[year_df['Pollutant'] == pollutant]
        
        for location in pollutant_df.columns[:-4]:
            if location != 'Date' and location != 'Time':
                max_pollution = pollutant_df[location].max()
        
                if pd.notnull(max_pollution):
                    max_timestamp = pollutant_df[(pollutant_df[location] == max_pollution)]['Time'].values[0]
                
                result.append({
                    'Year': year,
                    'Time': max_timestamp,
                    'Pollutant': pollutant,
                    'Location': location
                })

result_df = pd.DataFrame(result)
print(result_df)
print()
time_with_most_pollution = result_df['Time'].value_counts().idxmax()
print(f'The time of the year that seems to have the most pollution in Skopje is {time_with_most_pollution}')
```

# Question 4
What is the average pollution accross locations in every hour during the day?  .......... (2 points)

```python
# Solution
#Getting the daily data for each pollutant
df['Day'] = pd.to_datetime(df['Date']).dt.day
pm10_1h=df[df['Pollutant']=='PM10'].groupby(['Day','Year'])[locations].rolling(window=24, min_periods=16).mean().values
pm25_1h=df[df['Pollutant']=='PM25'].groupby(['Day','Year'])[locations].rolling(window=24, min_periods=16).mean().values
co_1h=df[df['Pollutant']=='CO'].groupby(['Day','Year'])[locations].rolling(window=8, min_periods=1).max().values
no2_1h=df[df['Pollutant']=='NO2'].groupby(['Day','Year'])[locations].rolling(window=8, min_periods=1).mean().values
o3_1h=df[df['Pollutant']=='O3'].groupby(['Day','Year'])[locations].rolling(window=1, min_periods=1).max().values
o3_1h= (24.45 *o3_1h)/48000
# this formula is used to transform the data form μg/m3 to ppb
# because in the sent documentation the AQI for O3, its index values and breakpoints are written for ppb

#Calcuating the AQI for every pollutant
import Subidx as s 
# Subidx is a module containing the functions to calculate the AQI

result_co = {location: [round(s.get_CO_subindex(x), 1) for x in co_1h[:, i]] for i, location in enumerate(locations)}
co_aqi = pd.DataFrame(result_co)
print(f'CO AQI:\n {co_aqi}')
print()

result_no2 = {location: [int(s.get_NO2_subindex(x)) for x in no2_1h[:, i]] for i, location in enumerate(locations)}
no2_aqi = pd.DataFrame(result_no2)
print(f'NO2 AQI:\n {no2_aqi}')
print()
      
result_o3 = {location: [round(s.get_O3_subindex(x), 3) for x in o3_1h[:, i]] for i, location in enumerate(locations)}
o3_aqi = pd.DataFrame(result_o3)
print(f'O3 AQI:\n {o3_aqi}')
print()
      
result_pm10 = {location: [int(s.get_PM10_subindex(x)) for x in pm10_1h[:, i]] for i, location in enumerate(locations)}
pm10_aqi = pd.DataFrame(result_pm10)
print(f'PM10 AQI:\n {pm10_aqi}')

result_pm25 = {location: [round(s.get_PM25_subindex(x), 1) for x in pm25_1h[:, i]] for i, location in enumerate(locations)}
pm25_aqi = pd.DataFrame(result_pm25)
print(f'PM25 AQI:\n{pm25_aqi}')
```

According to the AQI Technical Assistance Document the AQI of a location is equal to the maximum value of the AQI between the pollutants that are measured.

```python
aqi_df = [co_aqi, no2_aqi, o3_aqi, pm10_aqi, pm25_aqi]
aqi_values=np.round_(np.maximum.reduce(aqi_df), decimals=0)
aqi_values_df= pd.DataFrame(aqi_values, columns=locations)
print(f'The AQI values across locations in every hour during the day is:\n {aqi_values_df}')
```

## Questions 6
If the range of values for PM10 considered safe and unsafe is according to the following scale:

Good                                    0-50

Moderate 	                            51-154

Unhealthy for sensitive individuals 	155-254

Unhealthy 	                            255-354

Very unhealthy                          355-424

Hazardous                               425-504

On how many days in each year, was the average value of PM10 meauserd accross locations worse than moderate?  .......... (3 points)

```python
# Solution
avg_daily = df[df['Pollutant'] == 'PM10'].groupby(['Year', 'Date']).mean()

total_days =(avg_daily[locations] > 154).sum(axis=1).groupby('Year').apply(lambda x: (x > 0).sum())

print(f'Number of days for every year that the average value of PM10 accross all locations was worse than moderate:\n'
     f'{total_days}')
```

## Question 7
On how many days in each year, was the value of PM10 measured on at least one location worse than moderate? .......... (3 points)

```python
# Solution
daily_measure= df[df['Pollutant'] == 'PM10'].groupby(['Year', 'Date']).apply(lambda x: (x[locations] > 154).any(axis=1).any()).groupby('Year').sum()
print(f'Number of days for every year that the value of PM10 on at least on location was worse than moderate:\n'
     f'{daily_measure}')
```

## Question 8
Make a bar plot showing the average number of days per year when the measured value for PM10 was worse than moderate on at least one location.  .......... (3 points)

```python
# Solution
plt.figure(figsize = (12, 6))

moderate_days_avg = daily_measure.mean()

sns.barplot(x=daily_measure.index, y=daily_measure, color='#3274a1')
plt.axhline(y=moderate_days_avg, color='#e1812c', linestyle='--', label='Average Number of Days')
plt.xlabel('Year')
plt.ylabel('Number of Days')
plt.title('Average Number of Days per Year with PM10 Worse than Moderate').set_size(12)
plt.legend()
plt.show()
```
